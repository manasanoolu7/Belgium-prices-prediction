```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('clean_dataset.csv')
df
```

```python
df.describe()
```

```python
df['rooms_number'].unique()
```

```python
categorical_data=['house_is','property_subtype','equipped_kitchen_has','furnished','open_fire',
                 'terrace','garden','swimming_pool_has','region','building_state_agg']
numerical_data=['source','postcode','price','rooms_number','area','terrace_area',
               'garden_area','land_surface','facades_number',]
```

```python
for each_category in categorical_data:
    df[each_category]=df[each_category].astype('category')
    
```

```python
df.info()
```

```python
print(df['house_is'].unique())
print(df['property_subtype'].unique())
print(df['equipped_kitchen_has'].unique())
print(df['furnished'].unique())
print(df['open_fire'].unique())
print(df['terrace'].unique())
print(df['garden'].unique())
print(df['swimming_pool_has'].unique())
print(df['region'].unique())
print(df['building_state_agg'].unique())
```

```python
df['region'] = df['region'].cat.reorder_categories(['W', 'F', 'B'], ordered=True)
df['region'] = df['region'].cat.codes
print(df['region'])
```

```python
df['building_state_agg'] = df['building_state_agg'].cat.reorder_categories(['good', 'to_renovate', 'renovated'], ordered=True)
df['building_state_agg'] = df['building_state_agg'].cat.codes
print(df['building_state_agg'])
```

```python
df['property_subtype'].unique()
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python

df.drop(['facades_number'],inplace=True,axis=1)
```
